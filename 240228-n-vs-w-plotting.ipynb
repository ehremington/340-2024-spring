```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Let's start with a simpler plot and move upwards from there. 

For example I could let everything be 1 except $\omega$ and then add a complex number in there and look then.

```python
def a(w):
    return(1/(1-w**2+1j*w))

w = np.linspace(0.5,2,101)

fig0, ax0 = plt.subplots()
ax0.plot(w, np.real(a(w)))
ax0.plot(w, np.imag(a(w)))
```
