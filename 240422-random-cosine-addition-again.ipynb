# Plotting an exam question

I asked in number 3 whether a sum of cosines of random phase are equal to zero, so this is the answer to that question.

I previously investigated this problem in the fall 2020 version of this class. This was started based on a statement made in Pedrotti 3rd edition on pg 119. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 10000

n = np.random.random(N)
theta_i = n*np.pi*2

cos_theta = np.cos(theta_i)

cumulative_sum = np.cumsum(cos_theta)
```

```python
fig, ax = plt.subplots()

ax.plot(range(N),cumulative_sum)
```

# a little more

So I know that sum is not equal to 0 no matter how many terms you add up. BUT, the claim I made in class is that the average of many of those is equal to zero, so that is something I need to test.

```python
def generate_sum(N):
    n = np.random.random(N)
    theta_i = n*np.pi*2

    cos_theta = np.cos(theta_i)

    cumulative_sum = np.sum(cos_theta)
    return(cumulative_sum)

def many_sums(N):
    sum = 0
    for i in range(N):
        s = generate_sum(N)
        sum += s
    return(sum)
```

```python
many_sums(10000)
```

So that is not working out either. Number is not close to zero and never will be. BUT, what about that number in comparison to the other one. I mean the equation is:

$$E^2_0 = \sum_{i=1}^N E_{0i}^2 + 2\sum_{j>i}^N\sum_{i=1}^N E_{0i}E_{0j}\cos(\alpha_i-\alpha_j)$$

I have been letting $E_0 = 1$ so how big is that first term?

I mean up above that would be 10,000. That is the number of sums that I did. So considering that, this number is small.

Still though, both Pedrotti and Hecht give unsatisfying answers in the derivation of this. It is more that the number is small in comparison to the other, not that the number is zero. But I guess when we are talking about adding numbers up, it doesn't really matter how small the number is, what we really mean is their relative size.
